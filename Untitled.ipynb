```python
import datetime

x = [0.1, 0.05, 0.025, 0.0125, 0.00625, 0.003125, 0.00166]
y1 = []
y2 = []

for j in ["print", "!print"]:
    for i in x:
        
        begin =datetime.datetime.now()
        Testvalue = 0
        Steps = i
        Number = 5
        
        while Number + Steps > Testvalue**2:
            Testvalue += Steps
            if j == "print":
                print(Testvalue, Testvalue**2)
                
        end = datetime.datetime.now()
        cost = (end - begin).microseconds
        
        print (Testvalue, Testvalue**2)
        if j == "print":
            y1.append(cost)
        if j == "!print":
            y2.append(cost)
```

```python
plt.plot(x, y1, label = "Cost with print")
plt.plot (x, y2, label = "Cost without print")
plt.legend()
plt.xlabel("Accuracy")
plt.ylabel("Cost in microseconds")
plt.show()
```
